## Ideas for Lane Detection Pipeline

### Aim: Identify lanes and draw red overlays over the lanes.

Let us examine the data by plotting them. We are looking for hints to be able to identify the lanes. In order to do so, we follow the standard steps below:
<br>
1. Grayscale the image
2. Apply Gaussian blur
3. Apply Canny Transform
3. Choose Region of Interest
4. Get Hough lines
5. Calculate slope, intercept and lengths from hough lines
6. Plot the data for two test images
7. Pick up hints

```python
def get_image_characteristics(image):
    gray = grayscale(image)

    #Smoothening the edges helps remove jitters when finding canny edges
    g_blur = gaussian_blur(gray, 5)

    # Provide upper and lower threshold for canny Transforms (low_threshold: upper_threshold = 1: 3)
    low_threshold = 50
    high_threshold = 150
    edges = canny(g_blur, low_threshold, high_threshold)

    # Make vertices array for the Region of Interest
    imshape = image.shape
    vertices = np.array([[(0,imshape[0]),(420, 330), (520, 330), (imshape[1]*0.9,imshape[0]*0.9)]], dtype=np.int32)

    # Based on ROI select Region of Interest 
    masked_image = region_of_interest(edges, vertices)

    # Initialize parameters for extracting hough lines from the ROI image
    rho = 1 
    theta = np.pi/180 
    threshold = 35   
    min_line_length = 5
    max_line_gap = 5

    # Get hough lines 
    all_lines = hough_lines(masked_image, rho, theta, threshold, min_line_len=min_line_length, max_line_gap=max_line_gap)

    slopes = []
    intercepts = []
    line_lengths = []

    for line in all_lines:
        for x1, y1, x2, y2 in line:
            if x2 == x1:
                continue

            slope = (y2 - y1)/(x2 - x1)
            intercept = y1 - slope * x1
            euclidian_dist = np.sqrt((y2-y1)**2+(x2-x1)**2)
            slopes.append(slope)
            line_lengths.append(euclidian_dist)
            intercepts.append(intercept)

    return intercepts, slopes, line_lengths


image1 = mpimg.imread('test_images/solidWhiteCurve.jpg')
image2 = mpimg.imread('test_images/whiteCarLaneSwitch.jpg')
intercepts, slopes, line_lengths = get_image_characteristics(image1)
intercepts1, slopes1, line_lengths1 = get_image_characteristics(image2)

fig = plt.figure(figsize=(10, 12))

ax1 = fig.add_subplot(421)
ax1.imshow(image1)
ax1.set_title('White Lane')

ax2 = fig.add_subplot(422)
ax2.imshow(image2)
ax2.set_title('Yellow Lane')

ax3 = fig.add_subplot(423)
counts, bins, bars = ax3.hist(slopes, bins = 100)
ax3.set_title('Slope')
ax3.set_xlabel('Slopes')
ax3.set_ylabel('Count')

ax4 = fig.add_subplot(424)
counts, bins, bars = ax4.hist(slopes1, bins = 100)
ax4.set_title('Slope')
ax4.set_xlabel('Slopes')
ax4.set_ylabel('Count')

ax5 = fig.add_subplot(425)
counts, bins, bars = ax5.hist(line_lengths, bins = 100)
ax5.set_title('Line lengths')
ax5.set_xlabel('length')
ax5.set_ylabel('Count')

ax6 = fig.add_subplot(426)
counts, bins, bars = ax6.hist(line_lengths1, bins = 100)
ax6.set_title('Line lengths')
ax6.set_xlabel('length')
ax6.set_ylabel('Count')

ax7 = fig.add_subplot(427)
counts, bins, bars = ax7.hist(intercepts, bins = 100)
ax7.set_title('Intercept')
ax7.set_xlabel('length')
ax7.set_ylabel('Count')

ax8 = fig.add_subplot(428)
counts, bins, bars = ax8.hist(intercepts1, bins = 100)
ax8.set_title('Intercept')
ax8.set_xlabel('length')
ax8.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def convert_to_white(img, lower_threshold, upper_threshold):
    """
    Converts color content to white between hsv thresholds
    Returns an image with color between thresholds replaced with white
    """
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, lower_threshold, upper_threshold)    
    return cv2.addWeighted(img, 1,cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB), 1, 0)


def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    return cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)


def hough_lines_image(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def calc_weighted_average(values, weights):
    s = 0.0
    for x, y in zip(values, weights):
        s += x * y        
    return (s / sum(weights))

def get_lane_characteristics(lines):
    """
    'lines' should be the output of hough lines function
    
    Returns mean slope and intercepts of Left and Right lanes respectively
    """
    
    left_slopes = []
    left_intercepts = []
    right_slopes = []
    right_intercepts = []
    left_lengths = []
    right_lengths = []

    # Calculate slope and intercept of each line and segregate based on positive or negative slopes
    # Positive slopes being Right lane lines and Negative slopes for left lanes
    # Append the slopes and intecepts to lists
    for line in lines:
        for x1, y1, x2, y2 in line:
            if x2 == x1:
                continue
            
            slope = (y2 - y1)/(x2 - x1)
            intercept = y1 - slope * x1
            euclidian_dist = np.sqrt((y2-y1)**2+(x2-x1)**2)
            
            if slope < 0.5:
                left_slopes.append(slope)
                left_intercepts.append(intercept)
                left_lengths.append(euclidian_dist)
            elif slope >= 0.5:
                right_slopes.append(slope)
                right_intercepts.append(intercept)
                right_lengths.append(euclidian_dist)
            else:
                continue
                
    # Calculate the weight mean slope and intercept from the list
    # We will have a slope of intercept each for left and right lanes
    weighted_left_slope = calc_weighted_average(left_slopes, left_lengths)
    weighted_left_intercept = calc_weighted_average(left_intercepts, left_lengths)
    weighted_right_slope = calc_weighted_average(right_slopes, right_lengths)
    weighted_right_intercept = calc_weighted_average(right_intercepts, right_lengths)

    return weighted_left_slope, weighted_left_intercept, weighted_right_slope, weighted_right_intercept

def draw_lanes(img, horizon, left_slope, left_intercept, right_slope, right_intercept, line_width):
    """
    Provided slope and intercepts of Right and left lane, Image to overlay on
    
    Returns image merged with right and left lane lines
    """
    line_image = np.zeros_like(img)
    
    if np.isnan(right_slope) or np.isnan(left_slope) or np.isnan(right_intercept) or np.isnan(left_intercept):
        return img
    
    # Left lane
    bottom_left_y = img.shape[0]        
    bottom_left_x = int((bottom_left_y - left_intercept)/left_slope)
    
    top_left_y = horizon 
    top_left_x = (top_left_y - left_intercept)/left_slope 
    
    # Draw Left lane line
    cv2.line(line_image, (int(bottom_left_x), int(bottom_left_y)), (int(top_left_x), int(top_left_y)),  [255, 0, 0], line_width)
    
    # Right Lane
    bottom_right_y = img.shape[0]         
    bottom_right_x = int((bottom_right_y - right_intercept)/right_slope)
    
    top_right_y = horizon 
    top_right_x = (top_right_y - right_intercept)/right_slope 
    
    # Draw right lane line
    cv2.line(line_image, (int(bottom_right_x), int(bottom_right_y)), (int(top_right_x), int(top_right_y)),  [255, 0, 0], line_width)
    
    # img * α + img * β + λ
    return cv2.addWeighted(img, 1.0, line_image, 0.8, 0.0)
    
def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

```python
def paint_lanes(image):
    #Provide upper and lower hsv threshold of the color you want to replace with white
    yellow_lower = np.array([80, 50, 50]) 
    yellow_upper = np.array([100, 255, 255]) 
    
    #Replace the white strip on the lane to white strip for uniform performance
    whitened_image = convert_to_white(image, yellow_lower, yellow_upper)
    
    gray = grayscale(whitened_image)
    
    #Smoothening the edges helps remove jitters when finding canny edges
    g_blur = gaussian_blur(gray, 5)

    # Provide upper and lower threshold for canny Transforms (low_threshold: upper_threshold = 1: 3)
    low_threshold = 50
    high_threshold = 150
    edges = canny(g_blur, low_threshold, high_threshold)

    # Make vertices array for the Region of Interest
    imshape = image.shape
    vertices = np.array([[(0,imshape[0]),(420, 330), (520, 330), (imshape[1]*0.9,imshape[0]*0.9)]], dtype=np.int32)
    
    # Based on ROI select Region of Interest 
    masked_image = region_of_interest(edges, vertices)

    # Initialize parameters for extracting hough lines from the ROI image
    rho = 1 
    theta = np.pi/180 
    threshold = 35   
    min_line_length = 5
    max_line_gap = 5

    # Get hough lines 
    all_lines = hough_lines(masked_image, rho, theta, threshold, min_line_len=min_line_length, max_line_gap=max_line_gap)
    
    # Get average slopes for right and left lanes
    left_slope, left_intercept, right_slope, right_intercept = get_lane_characteristics(all_lines)
    
    # Horizon is the top y of the lane upto whice the lanes are to be drawn
    horizon = 330
    
    # Draw lanes on the orignal image based on slopes and intercepts of right and left lane
    return draw_lanes(image, horizon, left_slope, left_intercept, right_slope, right_intercept, 10)
```
